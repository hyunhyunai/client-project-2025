```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.style.use(['seaborn-v0_8-notebook'])
```

```python
np.random.seed(42)
data = {
    "year": np.random.choice([2023, 2024], size=100),
    "month":np.random.choice(range(1, 13), size=100),
    "total_sales":np.random.randint(1000, 5000, size=100)
}
df = pd.DataFrame(data)
df_2024 = df[df['year'] == 2024]

sorted_data = df_2024.sort_values(by='total_sales', ascending=False)
print(sorted_data.head(5))
    
```

```python
data = {
    "job": np.random.choice(["Engineer", "Designer", "Manager"], size=50),
    "salary":np.random.randint(4000, 12000, size=50),
}
df = pd.DataFrame(data)

avg_salary = df.groupby('job')['salary'].mean()
print(avg_salary)

highest_job = avg_salary.idxmax()
highest_salary = avg_salary.max()

print(f"직무: {highest_job}, 평균 급여: {highest_salary}")
```

```python
students = {
    "student_id": range(1, 21),
    "name": [f"Student_{i}" for i in range(1, 21)],
}
scores = {
    "student_id": range(1, 21),
    "math": np.random.randint(50, 100, size=20),
    "english": np.append(np.random.randint(50, 100, size=15), [None]*5),
    "science": np.random.randint(50, 100, size=20),
}
df_students = pd.DataFrame(students)
df_scores = pd.DataFrame(scores)

merged_data = pd.merge(df_students, df_scores, on='student_id', how='left')

clean_data = merged_data.dropna()
#max_scores = clean_data[['math','english','science']].max()
#print(max_scores)


clean_data = clean_data.copy()
clean_data['total_mean'] = clean_data[['math', 'english', 'science']].mean(axis=1)


sorted_data = clean_data.sort_values(by='total_mean', ascending = False)
print(sorted_data)

plt.figure(figsize=(10, 6))
plt.bar(sorted_data['name'], sorted_data['total_mean'], color='skyblue')
plt.title("Students' Total Mean Scores", fontsize=16)
plt.xlabel("Students", fontsize=12)
plt.ylabel("Total Mean Score", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
subject_means = clean_data[['math', 'english', 'science']].mean()

plt.figure(figsize=(6, 6))
plt.bar(subject_means.index, subject_means.values, color=['orange','green','blue'])
plt.title("Average Scores by Subject")
plt.ylabel("Average Score", fontsize=12)
plt.xlabel("Subject", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(clean_data['total_mean'], bins=10, color='purple', alpha=0.7)
plt.title("Distribution of Students' Total Mean Scores", fontsize=16)
plt.xlabel("Total Mean Score", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()
```

```python
data = {
    "order_id": range(1, 21),
    "total_sales": np.random.randint(100, 1000, size=20),
    "order_count": np.random.randint(1, 10, size=20),
}
df = pd.DataFrame(data)
df['average_order_value'] = df['total_sales'] / df['order_count']

top_10 = df.nlargest(10, 'average_order_value')
plt.bar(top_10['order_id'], top_10['average_order_value'])
plt.xlabel('Order ID')
plt.ylabel('Average Order Value')
plt.title('Top 10 Orders by Average Order Value')
plt.show()

```

```python
data = {
    "year": np.random.choice([2023, 2024], size=50),
    "product": np.random.choice(["Product_A", "Product_B", "Product_C"], size=50),
    "sales": np.random.randint(500, 5000, size=50),
}
df = pd.DataFrame(data)

pivot_table = df.pivot_table(values='sales', index='year', columns='product', aggfunc='sum')
print(pivot_table)

sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Sales Heatmap by Year and Product')
plt.show()
```

```python
data = {
    "hour": np.random.choice(range(24), size=100),
    "traffic": np.random.randint(100, 1000, size=100),
}
df = pd.DataFrame(data)

hourly_avg = df.groupby('hour')['traffic'].mean()
hourly_avg.plot(kind='bar', figsize=(14, 6))
plt.xlabel('Hour')
plt.ylabel('Average Traffic')
plt.xticks(rotation=0)
plt.title('Average Traffic by Hour')
plt.show()
```
