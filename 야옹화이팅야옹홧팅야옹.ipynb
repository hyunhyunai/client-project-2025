```python
df = pd.read_csv('The number of crimes in Russia 2008-2023.csv')
print(df.isnull().sum())
df = df.dropna()
```

```python
region_crime_rate = df.groupby('Region')['The number of crimes (cases per 100.000 population)'].mean()
top_10_regions = region_crime_rate.sort_values(ascending=False).head(10)
print(top_10_regions)

top_10_regions.plot(kind='bar', title='Top 10 Regions by Crime Rate',ylabel='Crime Rate (per 100,000 population)', xlabel='Region');
```

```python
"""crime_type = 'Murder'
murder_data = df[df['Type of crime'] == crime_type]
murder_by_year = murder_data.groupby('Year')['The number of crimes'].sum()
print(murder_by_year)"""
murder_data = df[df['Type of crime'] == 'Murder']
murder_by_region = murder_data.groupby('Region')['The number of crimes (cases per 100.000 population)'].mean()
top_10_murder_region = murder_by_region.sort_values(ascending=False).head(10)
print(top_10_murder_region)
top_10_murder_region.plot(kind='bar',title='top10Murderregion',ylabel='The number of crimes (cases per 100.000 population)', xlabel='Region')
```

```python
pivot_table = df.pivot_table(
    values='The number of crimes',
    index='Region',
    columns='Year',
    aggfunc='sum'
)
pivot_table = pivot_table.fillna(0)
print(pivot_table)
```
